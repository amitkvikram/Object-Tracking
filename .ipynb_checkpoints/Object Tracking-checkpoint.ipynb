<h1 align = "middle">Object Tracking</h1>
### Aim
In this notebook we implement a basic object tracking model. Goal is to track an object based on its colour information. 

### Procedure
First we will record a video, while recording we draw a rectangle around the object 
to be tracked using mouse. Then we press Esc key.After that a new window will open 
which will show the selected area. Then pressing Esc key closes current window and 
Three new window will open. One window is current recording, window named "mask" is our created
mask, third window named "Tracking" will show our tracking result.

### Theory

* The feature which we will use for tracking is color information. 
* When we select the object to be tracked(we will call it as roi(region of interest)) it is in 
  BGR. First we convert roi from BGR to HSV colorspace. We will discuss later why we convert
  colorspace to hsv, till then think it same as BGR.
* Next task is to find the color range to be tracked. For this we calculate mean and standard
  deviation of each channel(H,S,V) individually for roi. Then our color range is simply 
  ( [H_mean-H_std, H_mean+H_std], [S_mean-S_std, S_mean+S_std], [V_mean-V_std, V-mean+V_std] ).
* Let us define 
                lower_bound = (H_mean - H_std, S_mean - S_std, V_mean - V_std) and 
                upper_bound = (H_mean + H_std, S_mean + S_std, V_mean + V_std)
* Now to create mask we will use cv2.inRange function available in openCv library. cv2.inRange   function takes frame(current frame in video), lower_bound , upper_bound. It return an array 
  of same size as frame in which a pixel value in 255 if corresponding pixel value in frame 
  is in range [lower_bound,upper_bound] otherwise pixel value is zero.
* After creating mask, to track the object we take bitwise_and of frame and mask. When we take 
  bitwise and of frame and mask in our ouput we will get pixel value 0 where pixel value of
  mask is 0.Reason behind it is simple as each pixel is a 8-bit unsigned int, 0 means all 8 
  bits are zero, now when we take bitwise and of 0 with any bit result is zero.So all 8 bits
  in ouput, hence the pixel value will be 0. AND output pixel value will be same as pixel value 
  of frame where pixel value of mask is 255. Reason is same.
* In the window named "Tracking" we can see our tracking result.
* Why we need to change colorspace from BGR to HSV(or YCrCb)
    * Problem with BGR is mixing of chrominance(Color related information) and luminance( 
      Intensity related information) data. Suppose you have an image of a single-color plane with a shadow on it. In RGB colorspace, the shadow part will most likely have very different characteristics than the part without shadows. In HSV colorspace, the hue component of both patches is more likely to be similar. Since we are tracking object based on its color(and intensity) information, we should be able to distinguish color information from luminance.
    * In HSV three channel represents Hue(Primary Color or dominant wavelength), Saturation( shades of the color ) and value(intensity).So it separates the image intensity from the color information. Since we are trying to track object base on its color range this is really useful.YCrCb colorspace is also useful for same.
    

 

### 1.BGR

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(context = 'paper', style = 'white', rc={"font.size":8,"axes.titlesize":15,"axes.labelsize":12,
                              "axes.labelcolor":'r','xtick.labelsize':10,'axes.facecolor': '1',
                              'figure.facecolor':'1', "xtick.color":'0',"figure.figsize":(4,4),
                              'axes.edgecolor':'0','axes.grid':False,'axes.spines.bottom': False,
                              'axes.spines.left': False,'axes.spines.right': False, 'axes.spines.top': False,
                                               'text.color': 'b'}) 

def draw_rectangle(event, x, y, flag , param):
    global drawing, x_i, y_i, frame, flag1, x_j, y_j
    if event == cv2.EVENT_LBUTTONDOWN:
        flag1 = 0
        x_i, y_i = x, y
        drawing = True
    elif drawing==True and event == cv2.EVENT_MOUSEMOVE:
        x_j, y_j= x, y
        flag1 = 1
    elif flag1 == 1 and event ==cv2.EVENT_LBUTTONUP:
        drawing = False
        x_j, y_j = x, y
        

drawing = False
x_i, y_i, x_j, y_j = 0, 0, 0, 0
flag1 = 0

cap = cv2.VideoCapture()
if not cap.isOpened():
    cap.open(0)

ret, frame = cap.read()
cv2.imshow('frame', frame)
cv2.setMouseCallback('frame', draw_rectangle)

while cap.isOpened():
    ret, frame = cap.read()
    if flag1 == 1:
        cv2.rectangle(frame, (x_i, y_i), (x_j, y_j), (0,0, 255), 1)
        
    cv2.imshow('frame', frame)
    k = cv2.waitKey(1) & 0xFF
    if k==27 and flag1==1:
        break
    
cap.release()
cv2.destroyAllWindows()

roi = frame[y_i:y_j, x_i:x_j, :]
cv2.imshow('roi', roi)

if cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()

#############Plotting Histogram of ROI###############
fig = plt.figure(figsize = (10,10))
ax = plt.subplot(221, title="Histogram of Blue")
plt.xlabel("Value")
plt.ylabel("Count")
# sns.ylim(0, 255)
# sns.xlim(0, 255)
sns.distplot(roi[:,:,0].ravel(), ax =ax, color = 'r', bins = 256, kde_kws={'color':'k','lw':1})
ax = plt.subplot(222, title = "Histogram of Green")
plt.xlabel("Value")
plt.ylabel("Count")
sns.distplot(roi[:,:,1].ravel(), ax =ax, color = 'r',bins = 256, kde_kws={'color':'k','lw':1})
ax = plt.subplot(223, title = "Histogram of Red")
plt.xlabel("Value")
plt.ylabel("Count")
sns.distplot(roi[:,:,2].ravel(), ax =ax, color = 'r', bins = 256,kde_kws={'color':'k','lw':1})
ax = plt.subplot(224, title = "Object")
plt.imshow(roi)
plt.tight_layout()
plt.show()
######################################################

B = np.sum(roi[:,:,0])
G = np.sum(roi[:,:,1])
R = np.sum(roi[:,:,2])

B_mean = np.mean(roi[:,:,0])
G_mean = np.mean(roi[:, :, 1])
R_mean = np.mean(roi[:,:,2])

B_std = np.std(roi[:,:,0])
G_std = np.std(roi[:,:,1])
R_std = np.std(roi[:,:,2])

lower_bound = np.array([B_mean-B_std , G_mean-G_std, R_mean - R_std])
upper_bound = np.array([B_mean + B_std, G_mean + G_std, R_mean + R_std])

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    cap.open(0)

ret, frame = cap.read()
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    mask = cv2.inRange(frame, lower_bound, upper_bound)
    cv2.imshow('Mask', mask)
    res = cv2.bitwise_and(frame, frame, mask = mask)
    cv2.imshow('Tracking', res)
    if cv2.waitKey(1) & 0xFF == 27:
        break;

cap.release()
cv2.destroyAllWindows()
```

### 2. Converting from BGR to HSV(back to BGR)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(context = 'paper', style = 'white', rc={"font.size":8,"axes.titlesize":15,"axes.labelsize":12,
                              "axes.labelcolor":'r','xtick.labelsize':10,'axes.facecolor': '1',
                              'figure.facecolor':'1', "xtick.color":'0',"figure.figsize":(4,4),
                              'axes.edgecolor':'0','axes.grid':False,'axes.spines.bottom': False,
                              'axes.spines.left': False,'axes.spines.right': False, 'axes.spines.top': False,
                                               'text.color': 'b'}) 

def draw_rectangle(event, x, y, flag , param):
    global drawing, x_i, y_i, frame, flag1, x_j, y_j
    if event == cv2.EVENT_LBUTTONDOWN:
        flag1 = 0
        x_i, y_i = x, y
        drawing = True
    elif drawing==True and event == cv2.EVENT_MOUSEMOVE:
        x_j, y_j= x, y
        flag1 = 1
    elif flag1 == 1 and event ==cv2.EVENT_LBUTTONUP:
        drawing = False
        x_j, y_j = x, y
        

drawing = False
x_i, y_i, x_j, y_j = 0, 0, 0, 0
flag1 = 0

cap = cv2.VideoCapture()
if not cap.isOpened():
    cap.open(0)

ret, frame = cap.read()
cv2.imshow('frame', frame)
cv2.setMouseCallback('frame', draw_rectangle)

while cap.isOpened():
    ret, frame = cap.read()
    if flag1 == 1:
        cv2.rectangle(frame, (x_i, y_i), (x_j, y_j), (0,0, 255), 1)
        
    cv2.imshow('frame', frame)
    k = cv2.waitKey(1) & 0xFF
    if k==27 and flag1==1:
        break
    
cap.release()
cv2.destroyAllWindows()

roi = frame[y_i:y_j, x_i:x_j, :]
cv2.imshow('roi', roi)

if cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()

roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

#############Plotting Histogram of ROI###############
fig = plt.figure(figsize = (10,10))
ax = plt.subplot(221, title="Histogram of Hue")
plt.xlabel("Value")
plt.ylabel("Count")
# sns.ylim(0, 255)
# sns.xlim(0, 255)
sns.distplot(roi[:,:,0].ravel(), ax =ax, color = 'r', bins = 256, kde_kws={'color':'k','lw':1})
ax = plt.subplot(222, title = "Histogram of Saturation")
plt.xlabel("Value")
plt.ylabel("Count")
sns.distplot(roi[:,:,1].ravel(), ax =ax, color = 'r',bins = 256, kde_kws={'color':'k','lw':1})
ax = plt.subplot(223, title = "Histogram of Value")
plt.xlabel("Value")
plt.ylabel("Count")
sns.distplot(roi[:,:,2].ravel(), ax =ax, color = 'r', bins = 256,kde_kws={'color':'k','lw':1})
ax = plt.subplot(224, title = "Object")
plt.imshow(roi)
plt.tight_layout()
plt.show()
######################################################

H = np.sum(roi[:,:,0])
S = np.sum(roi[:,:,1])
V = np.sum(roi[:,:,2])

H_mean = np.mean(roi[:,:,0])
S_mean = np.mean(roi[:, :, 1])
V_mean = np.mean(roi[:,:,2])

H_std = np.std(roi[:,:,0])
S_std = np.std(roi[:,:,1])
V_std = np.std(roi[:,:,2])

lower_bound = np.array([H_mean-H_std , S_mean-S_std, V_mean - V_std])
upper_bound = np.array([H_mean + H_std, S_mean + S_std, V_mean + V_std])

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    cap.open(0)

ret, frame = cap.read()
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(frame, lower_bound, upper_bound)
#     frame = cv2.cvtColor(frame, cv2.COLOR_HSV2BGR)
    cv2.imshow('Mask', mask)
    res = cv2.bitwise_and(frame, frame, mask = mask)
    res = cv2.cvtColor(res, cv2.COLOR_HSV2BGR)
    cv2.imshow('Tracking', res)
    if cv2.waitKey(1) & 0xFF == 27:
        break;

cap.release()
cv2.destroyAllWindows()
```

### 3. Converting from BGR to YCrCb(back to BGR)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(context = 'paper', style = 'white', rc={"font.size":8,"axes.titlesize":15,"axes.labelsize":12,
                              "axes.labelcolor":'r','xtick.labelsize':10,'axes.facecolor': '1',
                              'figure.facecolor':'1', "xtick.color":'0',"figure.figsize":(4,4),
                              'axes.edgecolor':'0','axes.grid':False,'axes.spines.bottom': False,
                              'axes.spines.left': False,'axes.spines.right': False, 'axes.spines.top': False,
                                               'text.color': 'b'}) 

def draw_rectangle(event, x, y, flag , param):
    global drawing, x_i, y_i, frame, flag1, x_j, y_j
    if event == cv2.EVENT_LBUTTONDOWN:
        flag1 = 0
        x_i, y_i = x, y
        drawing = True
    elif drawing==True and event == cv2.EVENT_MOUSEMOVE:
        x_j, y_j= x, y
        flag1 = 1
    elif flag1 == 1 and event ==cv2.EVENT_LBUTTONUP:
        drawing = False
        x_j, y_j = x, y
        

drawing = False
x_i, y_i, x_j, y_j = 0, 0, 0, 0
flag1 = 0

cap = cv2.VideoCapture()
if not cap.isOpened():
    cap.open(0)

ret, frame = cap.read()
cv2.imshow('frame', frame)
cv2.setMouseCallback('frame', draw_rectangle)

while cap.isOpened():
    ret, frame = cap.read()
    if flag1 == 1:
        cv2.rectangle(frame, (x_i, y_i), (x_j, y_j), (0,0, 255), 1)
        
    cv2.imshow('frame', frame)
    k = cv2.waitKey(1) & 0xFF
    if k==27 and flag1==1:
        break
    
cap.release()
cv2.destroyAllWindows()

roi = frame[y_i:y_j, x_i:x_j, :]
cv2.imshow('roi', roi)

if cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()

roi = cv2.cvtColor(roi, cv2.COLOR_BGR2YCrCb)

#############Plotting Histogram of ROI###############
fig = plt.figure(figsize = (10,10))
ax = plt.subplot(221, title="Histogram of Y(luma)")
plt.xlabel("Value")
plt.ylabel("Count")
# sns.ylim(0, 255)
# sns.xlim(0, 255)
sns.distplot(roi[:,:,0].ravel(), ax =ax, color = 'r', bins = 256, kde_kws={'color':'k','lw':1})
ax = plt.subplot(222, title = "Histogram of Cr(chroma)")
plt.xlabel("Value")
plt.ylabel("Count")
sns.distplot(roi[:,:,1].ravel(), ax =ax, color = 'r',bins = 256, kde_kws={'color':'k','lw':1})
ax = plt.subplot(223, title = "Histogram of Cb(chroma)")
plt.xlabel("Value")
plt.ylabel("Count")
sns.distplot(roi[:,:,2].ravel(), ax =ax, color = 'r', bins = 256,kde_kws={'color':'k','lw':1})
ax = plt.subplot(224, title = "Object")
plt.imshow(roi)
plt.tight_layout()
plt.show()
######################################################

Y = np.sum(roi[:,:,0])
Cr = np.sum(roi[:,:,1])
Cb = np.sum(roi[:,:,2])

Y_mean = np.mean(roi[:,:,0])
Cr_mean = np.mean(roi[:, :, 1])
Cb_mean = np.mean(roi[:,:,2])

Y_std = np.std(roi[:,:,0])
Cr_std = np.std(roi[:,:,1])
Cb_std = np.std(roi[:,:,2])

lower_bound = np.array([Y_mean-Y_std , Cr_mean-Cr_std, Cb_mean - Cb_std])
upper_bound = np.array([Y_mean + Y_std, Cr_mean + Cr_std, Cb_mean + Cb_std])

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    cap.open(0)

ret, frame = cap.read()
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow('frame', frame)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2YCrCb)
    mask = cv2.inRange(frame, lower_bound, upper_bound)
    frame = cv2.cvtColor(frame, cv2.COLOR_YCrCb2BGR)
    res = cv2.bitwise_and(frame, frame, mask = mask)
    cv2.imshow('Mask', mask)
    cv2.imshow('Tracking', res)
    if cv2.waitKey(1) & 0xFF == 27:
        break;

cap.release()
cv2.destroyAllWindows()
```

### Conclusion
HSV and YCrCb both colorspace perform almost same. Also we don't see any significant improvement of changing BGR to HSV or YCrCb.
